# Projekt - Ingatlanfenntartási Bírságok Megértése és Előrejelzése

A város rendbontási bírságokat szab ki azon személyek számára, akik hagyják, hogy ingatlanjaik romos állapotban maradjanak. Minden évben a detroiti város millió dollárnyi bírságot szab ki a lakosokra, és minden évben sok ilyen bírság kifizetetlen marad. A kifizetetlen rendbontási bírságok behajtása költséges és fáradságos folyamat, ezért a város kíváncsi: hogyan növelhetjük a bírságok betartását?

Az első lépés ennek a kérdésnek a megválaszolásában az, hogy megértsük, mikor és miért nem fizethet meg egy lakos egy bírságot. Itt jön képbe az előrejelző modellezés. Ebben a feladatban az a feladatod, hogy előrejelezd, hogy egy adott rendbontási bírságot időben kifizetnek-e.

Egy adatfájlt áll a rendelkezésedre, amelyet a modellek betanításához és érvényesítéséhez használhatsz: train.csv. Ebben a fájlban minden sor egyetlen rendbontási bírságnak felel meg, és tartalmazza az információkat arról, hogy mikor, miért, és kinek szabták ki a bírságot. A célváltozó a compliance (megfelelés), amely Igaz, ha a bírságot korán, időben vagy a meghallgatás dátumától számított egy hónapon belül kifizették, Hamis, ha a bírságot a meghallgatás dátuma után vagy egyáltalán nem fizették ki, és null, ha a vétkes nem volt felelős. 

#### Fájlleírások (Csak ezeket az adatokat használd a modelled betanításához!)

- **train.csv** - az adathalmaz (minden 2004-2011 között kiadott bírság)
- **addresses.csv** & **latlons.csv** - a bírságok azonosítóinak címekhez való hozzárendelése, és a címek koordinátákkal való összekapcsolása. 
  **Megjegyzés:** A helytelenül írt címek hibásan geolokalizálhatók.

### Értékelés

Az előrejelzéseid értékelési szemponthja az, hogy mekkora a valószínűsége annak, hogy az adott rendbontási bírságot időben kifizetik.

Az értékelési metrika a feladat során az **AUC** (Receiver Operating Characteristic Curve alatti terület). 

A feladat során hozz létre egy függvényt, amely egy modellt betanít a rendbontási bírságok betartásának előrejelzésére a *train.csv* **train-test** bontásával képzett halmazok segítségével. Ezt a modellt használva térj vissza egy sorozattal, ahol az adatok az egyes bírságok időben történő kifizetésének valószínűségeit tartalmazzák, és az indexek a *ticket_id*-k. Ezután számítsd ki a ROC-görbe alatti területet az előrejelzéseid alapján. A cél legalább **0.75** értékű AUC. ((Aki a legmagasabb AUC-t eléri, pirospontot kap :))
