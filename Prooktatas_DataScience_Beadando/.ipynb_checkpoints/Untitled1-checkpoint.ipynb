```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", 200)

# 1) Beolvasás
def load_train(path="train.csv") -> pd.DataFrame:
    df = pd.read_csv(path, low_memory=False)
    return df

train = load_train("train.csv")

# 2) Alap oszlopok egységesítése (létezés ellenőrzés, hiányzók felvétele szükség szerint)
expected_num_cols = [
    "fine_amount", "late_fee", "discount_amount", "judgment_amount"
]
expected_date_cols = [
    "ticket_issued_date", "hearing_date", "payment_due_date", "payment_date"
]
id_col = "ticket_id"
target_col = "compliance"

for c in expected_num_cols:
    if c not in train.columns:
        train[c] = np.nan

for c in expected_date_cols:
    if c not in train.columns:
        train[c] = pd.NaT

if id_col not in train.columns:
    raise ValueError("Hiányzik a ticket_id oszlop a train.csv-ből.")

if target_col not in train.columns:
    raise ValueError("Hiányzik a compliance oszlop a train.csv-ből.")

# 3) Típuskonverziók
# ID-k stringre
train[id_col] = train[id_col].astype(str)

# Pénzügyi oszlopok numerikusra
for c in expected_num_cols:
    train[c] = pd.to_numeric(train[c], errors="coerce")

# Dátumok datetime-ra
for c in expected_date_cols:
    train[c] = pd.to_datetime(train[c], errors="coerce", utc=True)

# 4) Compliance tisztítás (csak True/False marad)
def normalize_compliance(series: pd.Series) -> pd.Series:
    s = series.copy()
    # String normalizálás
    if s.dtype == object:
        s = s.str.strip().str.lower()
        s = s.replace({
            "true": True, "false": False,
            "1": True, "0": False,
            "yes": True, "no": False
        })
    # Numerikus -> bool
    if pd.api.types.is_numeric_dtype(s):
        s = s.astype("float")
        s = s.map(lambda x: True if x == 1 else (False if x == 0 else np.nan))
    return s.astype("object")

train[target_col] = normalize_compliance(train[target_col])

# Csak valódi True/False sorok megtartása
train = train[train[target_col].isin([True, False])].copy()

# 5) Hibás/irreális értékek kezelése pénzügyekben
for c in expected_num_cols:
    train.loc[train[c] < 0, c] = np.nan

# Hiányzók kitöltése 0-val a díjtételeknél (minimál baseline)
fill_zero_cols = ["late_fee", "discount_amount"]
for c in fill_zero_cols:
    train[c] = train[c].fillna(0.0)

for c in ["fine_amount", "judgment_amount"]:
    median_val = train[c].median(skipna=True)
    train[c] = train[c].fillna(median_val)

# 6) Extrém outlierek visszavágása (winsorization-hoz hasonló egyszerű cap)
def cap_outliers(s: pd.Series, lower_q=0.001, upper_q=0.999) -> pd.Series:
    low = s.quantile(lower_q)
    high = s.quantile(upper_q)
    return s.clip(lower=low, upper=high)

for c in expected_num_cols:
    if train[c].notna().any():
        train[c] = cap_outliers(train[c])

# 7) Dátum-logika minimális ellenőrzés
min_date = pd.Timestamp("1990-01-01", tz="UTC")
max_date = pd.Timestamp("2035-01-01", tz="UTC")
for c in expected_date_cols:
    mask_bad = (train[c].notna()) & ((train[c] < min_date) | (train[c] > max_date))
    train.loc[mask_bad, c] = pd.NaT

# 8) Kimenet előkészítése a következő lépésekhez
clean_columns = [id_col, target_col] + expected_num_cols + expected_date_cols
clean_columns = [c for c in clean_columns if c in train.columns]
clean_train = train[clean_columns].reset_index(drop=True)

# Ellenőrzésként visszaadandó objektumok (ha függvénybe tesszük)
def get_clean_train() -> pd.DataFrame:
    return clean_train

# Rövid diagnosztika
print("Sorok száma a tisztítás után:", len(clean_train))
print("Pozitív arány (True):", clean_train[target_col].mean())
print(clean_train.dtypes)
```

```python
def load_addresses(path="addresses.csv") -> pd.DataFrame:
    df = pd.read_csv(path, low_memory=False)
    # Normalizáljuk a kulcsokat: ticket_id string, address oszlop létezzen
    if "ticket_id" not in df.columns:
        raise ValueError("addresses.csv: hiányzik a ticket_id oszlop.")
    if "address" not in df.columns:
        raise ValueError("addresses.csv: hiányzik az address oszlop.")
    df["ticket_id"] = df["ticket_id"].astype(str).str.strip()
    df["address"] = df["address"].astype(str).str.strip()
    
    df = df.drop_duplicates(subset=["ticket_id"], keep="first")
    return df

def load_latlons(path="latlons.csv") -> pd.DataFrame:
    df = pd.read_csv(path, low_memory=False)
    
    expected_cols = {"address", "lat", "lon"}
    missing = expected_cols - set(df.columns)
    if missing:
        raise ValueError(f"latlons.csv: hiányzó oszlop(ok): {missing}")
    
    df["address"] = df["address"].astype(str).str.strip()
    
    df["lat"] = pd.to_numeric(df["lat"], errors="coerce")
    df["lon"] = pd.to_numeric(df["lon"], errors="coerce")
    
    df = df.drop_duplicates(subset=["address"], keep="first")
    
    lat_ok = df["lat"].between(41.0, 45.0, inclusive="both")
    lon_ok = df["lon"].between(-86.0, -82.0, inclusive="both")
    df.loc[~(lat_ok & lon_ok), ["lat", "lon"]] = np.nan  # hibás geokódokat NaN-re tesszük
    return df

addresses = load_addresses("addresses.csv")
latlons = load_latlons("latlons.csv")

# 1) train + addresses join ticket_id alapján
train_addr = clean_train.merge(addresses, on="ticket_id", how="left")

# 2) cím + latlon join address alapján
train_geo = train_addr.merge(latlons, on="address", how="left")

# 3) Duplikációk ellenőrzése: egy ticket_id-hez egy sor
if train_geo.duplicated(subset=["ticket_id"]).any():
    
    train_geo["_has_geo"] = train_geo["lat"].notna() & train_geo["lon"].notna()
    train_geo = (train_geo
                 .sort_values(by=["ticket_id", "_has_geo"], ascending=[True, False])
                 .drop_duplicates(subset=["ticket_id"], keep="first")
                 .drop(columns="_has_geo"))

# 4) Kimenet a következő lépéshez (feature engineering)
geo_columns = ["ticket_id", "compliance", "lat", "lon", "address"]
keep_cols = [c for c in geo_columns if c in train_geo.columns]
train_with_geo = train_geo[keep_cols + [c for c in clean_train.columns if c not in ["ticket_id", "compliance"]]].copy()

# Diagnosztika
print("Összekapcsolás után sorok:", len(train_with_geo))
print("Koordinátával rendelkező arány:", train_with_geo["lat"].notna().mean())
```

```python
def make_basic_features(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy()

    # 1) Időjellemzők kinyerése
    date_cols = ["ticket_issued_date", "hearing_date", "payment_due_date"]
    for c in date_cols:
        if c in out.columns:
            # hónap és év
            out[f"{c}_year"] = out[c].dt.year
            out[f"{c}_month"] = out[c].dt.month

    # 2) Egyszerű időkülönbségek (napok)
    if {"ticket_issued_date", "hearing_date"}.issubset(out.columns):
        out["days_issue_to_hearing"] = (out["hearing_date"] - out["ticket_issued_date"]).dt.days

    if {"ticket_issued_date", "payment_due_date"}.issubset(out.columns):
        out["days_issue_to_due"] = (out["payment_due_date"] - out["ticket_issued_date"]).dt.days

    # 3) Pénzügyi transzformációk: log1p az erősen ferde eloszlás miatt
    money_cols = ["fine_amount", "late_fee", "discount_amount", "judgment_amount"]
    for c in money_cols:
        if c in out.columns:
            out[f"log1p_{c}"] = np.log1p(out[c].clip(lower=0))

    # 4) Egyszerű arányok, különbségek
    if {"judgment_amount", "fine_amount"}.issubset(out.columns):
        
        out["judgment_minus_fine"] = out["judgment_amount"] - out["fine_amount"]
        out["judgment_over_fine"] = np.where(out["fine_amount"] > 0,
                                             out["judgment_amount"] / out["fine_amount"],
                                             np.nan)

    # 5) Helyadatok baseline-ként
    for c in ["lat", "lon"]:
        if c not in out.columns:
            out[c] = np.nan

    # 6) Kategóriák minimális kezelése (opcionális baseline)
    
    if "disposition" in out.columns:
        topk = out["disposition"].value_counts().nlargest(5).index
        for val in topk:
            out[f"disp_is_{str(val)[:20].replace(' ','_').lower()}"] = (out["disposition"] == val).astype(int)

    # 7) Modellezésre szánt feature lista összeállítása
    base_feats = [
        # Pénzügyek (nyers + log)
        "fine_amount", "late_fee", "discount_amount", "judgment_amount",
        "log1p_fine_amount", "log1p_late_fee", "log1p_discount_amount", "log1p_judgment_amount",
        # Idő
        "ticket_issued_date_year", "ticket_issued_date_month",
        "hearing_date_year", "hearing_date_month",
        "payment_due_date_year", "payment_due_date_month",
        "days_issue_to_hearing", "days_issue_to_due",
        # Hely
        "lat", "lon",
    ]
    
    cat_feats = [c for c in out.columns if c.startswith("disp_is_")]
    features = [f for f in base_feats + cat_feats if f in out.columns]

    # 8) Kimenet: id, target, features
    cols_out = ["ticket_id", "compliance"] + features
    cols_out = [c for c in cols_out if c in out.columns]
    return out[cols_out].copy()

features_df = make_basic_features(train_with_geo)

print("Feature mátrix méret:", features_df.shape)
print("Példa oszlopok:", features_df.columns[:15].tolist())
```

```python
from sklearn.model_selection import train_test_split

def make_train_test(features_df: pd.DataFrame, test_size: float = 0.25, seed: int = 42):
    df = features_df.dropna(subset=["compliance"]).copy()

    # Cél és azonosítók
    y = df["compliance"].astype(int).values
    ids = df["ticket_id"].astype(str).values

    # X kiválasztása: minden, ami nem id és nem target
    drop_cols = {"ticket_id", "compliance"}
    feature_cols = [c for c in df.columns if c not in drop_cols]

    # Csak numerikus oszlopok megtartása (egyszerű baseline döntés)
    num_cols = [c for c in feature_cols if pd.api.types.is_numeric_dtype(df[c])]
    X = df[num_cols].values

    # Train-test split stratifikációval
    X_train, X_test, y_train, y_test, ids_train, ids_test = train_test_split(
        X, y, ids, test_size=test_size, random_state=seed, stratify=y
    )

    out = {
        "X_train": X_train,
        "X_test": X_test,
        "y_train": y_train,
        "y_test": y_test,
        "ids_train": ids_train,
        "ids_test": ids_test,
        "feature_names": num_cols,
    }
    return out

split_data = make_train_test(features_df, test_size=0.25, seed=42)

print("Train méret:", split_data["X_train"].shape, "Test méret:", split_data["X_test"].shape)
print("Pozitív arány train/test:", split_data["y_train"].mean(), split_data["y_test"].mean())
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score


model = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),   # NaN -> oszlop-medián
    ("scaler", StandardScaler(with_mean=True, with_std=True)),
    ("clf", LogisticRegression(
        C=1.0,
        class_weight="balanced",
        solver="lbfgs",
        max_iter=1000,
        random_state=42
    ))
])

model.fit(split_data["X_train"], split_data["y_train"])
probs_test = model.predict_proba(split_data["X_test"])[:, 1]
print("Példa valószínűségek (első 5):", np.round(probs_test[:5], 4))

auc = roc_auc_score(split_data["y_test"], probs_test)
print("Teszt AUC:", round(auc, 4))
```

```python
from sklearn.metrics import roc_auc_score, roc_curve

# 1) Előrejelzési valószínűségek a pozitív osztályra
probs_test = model.predict_proba(split_data["X_test"])[:, 1]

# 2) ROC-AUC számítás
auc = roc_auc_score(split_data["y_test"], probs_test)
print("Teszt ROC-AUC:", round(auc, 4))

# 3) Kötelező kimenet előállítása: Pandas Series (valószínűségek), ticket_id index
pred_series = pd.Series(probs_test, index=split_data["ids_test"])
pred_series.index.name = "ticket_id"
pred_series.name = "compliance_probability"


fpr, tpr, thresholds = roc_curve(split_data["y_test"], probs_test)

print("Előrejelzések száma:", pred_series.shape)
print("Példa (első 5):")
print(pred_series.head())
```
